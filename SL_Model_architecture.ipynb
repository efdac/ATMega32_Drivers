```python
from sklearn.model_selection import train_test_split
from sklearn import datasets

X_train, X_test, y_train, y_test = train_test_split(x, y)
print('Training Images Count :' ,len(X_train),'Training Labels Count:',len(y_train))
```

```python
print('testing Images Count :' ,len(X_test),'testing Labels Count:',len(y_test))
```

```python
# Basic CNN Model
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(len(labels), activation ='softmax'))
model.summary()
```

```python
X_train=np.asarray(X_train)
X_test=np.asarray(X_test)
y_train=np.asarray(y_train)
y_test=np.asarray(y_test)
```

```python
type(X_train).shape,type(X_test).shape,type(y_train.shape),type(y_test.shape)
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
history = model.fit(np.array(X_train), y_train, epochs=20, validation_data=(np.array(X_test), y_test)
,shuffle=True)
```
